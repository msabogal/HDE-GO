```python
from classy import Class
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
paramsGO = {
    'H0': 67.69 , 
    'T_cmb': 2.725,
    'omega_b': 0.02245,
    'omega_cdm': 0.1197,
    'n_s': 0.9718,
    'ln10^{10}A_s': 3.0335,
    'tau_reio': 0.0508,
    'Omega_Lambda': 0.,
    'Omega_k': 0.,
    'fluid_equation_of_state': 'GO',
    'alpha_GO': 0.96,
    'beta_GO': 0.48,
    'gauge': 'newtonian', 
    'use_ppf': 'no',
    'P_k_max_1/Mpc':3.0,
    'k_output_values': '0.0001,0.01, 0.1',
    'evolver': 0, 
    'output': 'tCl,mPk'}
```

```python
paramsLCDM = {
    'H0': 67.69 , 
    'T_cmb': 2.725,
    'omega_b': 0.02245,
    'omega_cdm': 0.1197,
    'n_s': 0.9718,
    'ln10^{10}A_s': 3.0335,
    'tau_reio': 0.0508,
    'Omega_fld': 0.,
    'Omega_k': 0.,
    'gauge': 'newtonian', 
    'P_k_max_1/Mpc':3.0,
    'k_output_values': '0.0001,0.01, 0.1',
    'output': 'tCl,mPk'}
```

```python
cosmo = Class() ; lCDM = Class() 
```

```python
cosmo.set(paramsGO); lCDM.set(paramsLCDM)
```

```python
cosmo.compute()
```

```python
lCDM.compute()
```

```python
Background_GO=cosmo.get_background();
Background_lCDM=lCDM.get_background(); 
```

```python
z=Background_GO["z"]; H=Background_GO["H [1/Mpc]"]; 
zlcdm=Background_lCDM["z"]; Hlcdm=Background_lCDM["H [1/Mpc]"]
```

```python
matplotlib.rcParams['xtick.direction'] = "in"
matplotlib.rcParams['ytick.direction'] = "in"

fig=plt.figure(figsize=(9,7))
plt.figure(1)
plot(z,H,'b-',linewidth=2)
plot(zlcdm,Hlcdm,'k--',linewidth=2.5)
plt.xlabel(r'$z$')
plt.ylabel(r'$H\,(z)\,\,\,[1/Mpc]$')
yscale('log')
xscale('log')
#plt.ylim(0,0.25)
#plt.xlim(0,140)
plt.title(r'$\alpha = 0.96 \,\,\,\, \beta = 0.48 $',size=14)
plt.legend(labels=['GO', r'$\Lambda$CDM'],loc=5,prop={'size': 14})
plt.rcParams.update({'font.size': 14})
plt.tick_params(direction='in', length=4, width=2, colors='k', right=True, top=True, 
                labelright='off',labeltop='off')

ax = fig.add_axes([0.215, 0.55, 0.35, 0.3])
ax.plot(z,H,'b-',linewidth=2)
ax.plot(zlcdm,Hlcdm,'k--',linewidth=2.5)
ax.set_xlabel(r'$z$', fontsize = 10)
ylabel(r'$H\,(z)\,\,\,[1/Mpc]$',size=10)
ylim(0,0.25)
xlim(0,140)
ax.xaxis.set_tick_params(labelsize=9)
ax.yaxis.set_tick_params(labelsize=9)
tick_params(direction='in', length=4, width=2, colors='k', right=True, top=True, 
                labelright='off',labeltop='off')

plt.savefig("HvsZ_GO.pdf", format="pdf", dpi=20000)
show()
```

```python
DeltaE=(H/Hlcdm - 1)*100; DeltaElcdm=(Hlcdm/Hlcdm - 1)*100; 
```

```python
plt.figure(figsize=(9,7))
plot(z,DeltaE,'b-',linewidth=2)
#plot(zlcdm,DeltaElcdm,'k--',linewidth=2)
plt.xlabel(r'$z$')
plt.ylabel(r'$\Delta E\,(z)$')
yscale('log')
xscale('log')
plt.title(r'$\alpha = 0.96 \,\,\,\, \beta = 0.48 $',size=14)
plt.tick_params(direction='in', length=4, width=2, colors='k', right=True, top=True, 
                labelright='off',labeltop='off')
plt.savefig("DeltaE.pdf", format="pdf", dpi=2000)
```

```python
w=Background_GO["(.)w_fld"];
wlcdm=-Background_lCDM["(.)rho_lambda"]/Background_lCDM["(.)rho_lambda"]; 
```

```python
plt.figure(figsize=(9,7))
plot(z,w,'b-',linewidth=2.5)
plot(zlcdm,wlcdm,'k--',linewidth=2.5)
plt.xlabel(r'$z$')
plt.ylabel(r'$w_{DE}\,(z)$')
xscale('log')
plt.legend(labels=['GO', r'$\Lambda$CDM'],loc=7,prop={'size': 14})
plt.title(r'$\alpha = 0.96 \,\,\,\, \beta = 0.48 $',size=14)
plt.tick_params(direction='in', length=4, width=2, colors='k', right=True, top=True, 
                labelright='off',labeltop='off')
plt.savefig("w_GO.pdf", format="pdf", dpi=2000)
```

```python
OmegaDE=Background_GO["(.)rho_fld"]/Background_GO["(.)rho_crit"]; 
OmegaLCDM=Background_lCDM["(.)rho_lambda"]/Background_lCDM["(.)rho_crit"]; 
```

```python
plt.figure(figsize=(9,7))
plot(z,OmegaDE,'b-',linewidth=2)
plot(zlcdm,OmegaLCDM,'k--',linewidth=2)
plt.xlabel(r'$z$')
plt.ylabel(r'$\Omega_{DE}\,(z)$')
xscale('log')
plt.legend(labels=['GO', r'$\Lambda$CDM'],loc=1,prop={'size': 14})
plt.title(r'$\alpha = 0.96 \,\,\,\, \beta = 0.48 $',size=14)
plt.tick_params(direction='in', length=4, width=2, colors='k', right=True, top=True, 
                labelright='off',labeltop='off')
plt.savefig("Omega_GO.pdf", format="pdf", dpi=2000)
```

```python
Cl_GO= cosmo.raw_cl();l=Cl_GO["ell"]; Cl=Cl_GO["tt"]; 
```

```python
Cl_lcdm= lCDM.raw_cl(); l2=Cl_lcdm["ell"]; Cl2=Cl_lcdm["tt"]; 
```

```python
plt.figure(figsize=(9,7))
plt.figure(1)
plot(l,l*(l+1.)/2/pi*Cl,'b-')
plot(l2,l2*(l2+1.)/2/pi*Cl2,'k-.')
plt.xlim(2,2500)
plt.xlabel(r'$\ell$')
plt.ylabel(r'$[\ell(\ell+1)/2\pi]  C_\ell^\mathrm{TT}$')
yscale('log'); xscale('log')
plt.legend(labels=['GO', r'$\Lambda$CDM'],loc=2,prop={'size': 14})
plt.title(r'$\alpha = 0.96 \,\,\,\, \beta = 0.48 $',size=14)
plt.tick_params(direction='in', length=4, width=2, colors='k', right=True, top=True, 
                labelright='off',labeltop='off')
plt.savefig("cl_GO.pdf", format="pdf", dpi=2000)
```

```python
kk = np.logspace(-6,np.log10(3),1000) # k in h/Mpc
Pk = []; Pklcdm = [] # P(k) in (Mpc/h)**3
h = cosmo.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk:
    Pk.append(cosmo.pk(k*h,0.)*h**3) # function .pk(k,z)
    Pklcdm.append(lCDM.pk(k*h,0.)*h**3)
```

```python
# plot P(k)
plt.figure(figsize=(9,7))
plt.xscale('log');plt.yscale('log'); #plt.xlim(kk[0],kk[-1])
plt.xlabel(r'$k \,\,\,\, [h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k) \,\,\,\, [\mathrm{Mpc}/h]^3$')
plt.plot(kk,Pk,'b-')
plt.plot(kk,Pklcdm,'k-.')
plt.legend(labels=['GO', r'$\Lambda$CDM'],prop={'size': 14})
plt.title(r'$\alpha = 0.96 \,\,\,\, \beta = 0.48 $',size=14)
plt.tick_params(direction='in', length=4, width=2, colors='k', right=True, top=True, 
                labelright='off',labeltop='off')
plt.savefig("pk_GO.pdf", format="pdf", dpi=2000)
```
